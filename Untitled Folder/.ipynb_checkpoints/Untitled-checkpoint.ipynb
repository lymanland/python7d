```python
import matplotlib.pyplot as plt
import numpy as np

# 从[-1,1]中等距去50个数作为x的取值
x = np.linspace(-1, 1, 50)
print(x)
# 使用figure()函数重新申请一个figure对象
# 注意，每次调用figure的时候都会重新申请一个figure对象
plt.figure()
# 第一个是横坐标的值，第二个是纵坐标的值
plt.plot(x)

y=2*x+1
# 第一个参数表示的是编号，第二个表示的是图表的长宽
plt.figure(num = 3, figsize=(8, 5))
plt.plot(x,y)

plt.plot(x, y, 
         color='red',   # 线颜色
         linewidth=1.0,  # 线宽 
         linestyle='--'  # 线样式
        )
plt.xlabel("I am x")
plt.ylabel("I am y")

#[<matplotlib.lines.Line2D at 0x6f6da81a58>]
plt.show()

```
